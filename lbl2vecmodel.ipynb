```python
import pandas as pd
from lbl2vec import Lbl2Vec
from gensim.utils import simple_preprocess
from gensim.models.doc2vec import Doc2Vec, TaggedDocument
from gensim.parsing.preprocessing import strip_tags
from sklearn.metrics import f1_score
```

```python
drawings = pd.read_csv(r"C:\Users\gramos1\OneDrive - Reworld\DrawingNames.csv",sep=',',header=0, names = ['DrawingName'])
labels = pd.read_csv(r"C:\Users\gramos1\OneDrive - Reworld\drawinglabels.csv", sep=',')
labels = labels.rename(columns={"Unnamed: 0": "Category"})
labels['keywords'] = labels['keywords'].apply(lambda x: x.split())
labels['keywords'] = labels['keywords'].apply(lambda description_keywords : [keyword.lower() for keyword in description_keywords])
labels['number_of_keywords'] = labels['keywords'].apply(lambda row: len(row))
```

```python
def tokenize(doc): # convert document into list of lowercase tokens
    #Need to also make it so that the first token gets 
    return simple_preprocess(strip_tags(doc),min_len = 1,max_len =50)
```

```python
drawings['tagged_docs'] = drawings.apply(lambda row: TaggedDocument(tokenize(row['DrawingName']), [str(row.name)]), axis=1)
```

```python
lbl2vec_model = Lbl2Vec(keywords_list = list(labels['keywords']),
                        tagged_documents=drawings['tagged_docs'],
                        label_names = list(labels['Categories']),
                        epochs = 15,
                        min_count = 1,
                        similarity_threshold = 0.50,
                        min_num_docs=1,
                        verbose = True,
                       )
```

```python
lbl2vec_model.fit()
```

```python
model_docs_lbl_similarities = lbl2vec_model.predict_model_docs()
```
